```python
import pandas as pd
```

```python
df = pd.read_csv('../sept9_syndirella_final_output.csv')
df.columns
```

```python
car = df[df['num_steps'] > 0]
car = car[~car['error_type'].notna()]
car = car[['smiles', '1_r1_smiles', '1_r2_smiles', '1_reaction', '2_r1_smiles',
       '2_r2_smiles', '2_reaction', '3_r1_smiles', '3_reaction', '4_r1_smiles',
       '4_r2_smiles', '4_reaction', 'num_steps', 'hit1_short', 'hit2_short', 'inchi_key', 'compound_set', 'total_num_unique_products', 'total_num_products_enumstereo', 'num_placed', 'num_successful']]
car
```

```python
non_car = df[df['error_type'] == 'NoSynthesisRoute']
non_car = non_car[['smiles', 'hit1_short', 'hit2_short', 'inchi_key', 'compound_set', 'error_type']]
non_car
```

```python
car.to_csv('sept10_syndirella_CHIKV_MAC_car_routes.csv', index=False)
non_car.to_csv('sept10_syndirella_CHIKV_MAC_no_car_routes.csv', index=False)
```
